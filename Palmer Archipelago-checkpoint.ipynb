# Cleaning The Dataset

```python
# The entire row of index 3 and 339 have null values
#The are only 2 non-null values for Comments, the Region(Anvers) is the same throughout the dataset,
#Sample number is same as index number but starts from 1

penguin.drop(penguin.index[[3,339]],axis = 0,inplace = True)
penguin.drop(['Comments','Region','Sample Number'],axis = 1,inplace = True)
penguin['Sex'].fillna(penguin['Sex'].mode()[0],inplace = True)
penguin['Delta 15 N (o/oo)'].fillna(penguin['Delta 15 N (o/oo)'].mean(),inplace = True)
penguin['Delta 13 C (o/oo)'].fillna(penguin['Delta 13 C (o/oo)'].mean(),inplace = True)
```

# Observations of Penguins by Species, Islands, Body Mass and Sex

```python
plt.rcParams['figure.figsize'] = (10,8)
plt.figure(figsize = (15,10))
plt.subplot(1,2,1)
sns.barplot(x = 'species',y='body_mass_g',hue = 'sex',data = size)
plt.title('Body Mass of Species by Sex')

plt.subplot(1,2,2)
sns.countplot(x ='species',hue = 'island',data = size)
plt.title('Which Penguin Species belong to Which Island')

plt.legend()
plt.tight_layout()
plt.show()
```

Heaviest species for both sexes is Gentoo

Lightest species for sexes is Chinstrap

Adelie belongs is found in Torgersen, Biscoe and Dream

Chinstrap is only found in Dream

Gentoo is only found in Biscoe
